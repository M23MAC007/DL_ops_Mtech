```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = [-3.5, -1.2, 0, 2.8, -4.1, 1.5, -0.7, 3.2, -2.4, 4.6]
y = []

for val in x:
    y.append(sigmoid(val))

plt.plot(x, y, marker='o')  
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function')
plt.grid(True)
plt.show()

```

```python

def relu(x):

    return np.maximum(0,x)


x = [-3.5, -1.2, 0, 2.8, -4.1, 1.5, -0.7, 3.2, -2.4, 4.6]
y = []

for val in x:
    y.append(relu(val))

plt.plot(x,y)
plt.xlabel('x') 
plt.ylabel('relu')
plt.title('ReLU Function')
plt.grid(True)
plt.show()   
```

```python
def leaky_relu(x,alpha=0.05):
    return np.where(x>0,x,alpha*x)

x = [-3.5, -1.2, 0, 2.8, -4.1, 1.5, -0.7, 3.2, -2.4, 4.6]
y = []

for val in x:
    y.append(leaky_relu(val))


plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('leaky_Relu')
plt.title('Leaky_Relu Function')
plt.grid(True) 
plt.show()  
```

```python
def Tanh(x):
    return np.tanh(x)

x = [-3.5, -1.2, 0, 2.8, -4.1, 1.5, -0.7, 3.2, -2.4, 4.6]
y = []

for val in x:
    y.append(Tanh(val))


plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('Tanh')
plt.title('Tanh Function')
plt.grid(True) 
plt.show()  

    
```
