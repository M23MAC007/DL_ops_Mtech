```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)
y = sigmoid(x)

# Plot the sigmoid function
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
# Define ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate x values
x = np.linspace(-10, 10, 100)
y = relu(x)

# Plot the ReLU function
plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.show()
```

```python
# Define Leaky ReLU function
def leaky_relu(x, alpha=0.05):
    return np.where(x > 0, x, alpha * x)

# Generate x values
x = np.linspace(-10, 10, 100)
y = leaky_relu(x)

# Plot the Leaky ReLU function
plt.plot(x, y)
plt.title('Leaky ReLU Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.show()
```

```python
# Generate x values
x = np.linspace(-5, 5, 100)

# Compute y values using the tanh function
y = np.tanh(x)

# Plot the tanh function
plt.plot(x, y)
plt.title('Hyperbolic Tangent Function (tanh)')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.show()
```
